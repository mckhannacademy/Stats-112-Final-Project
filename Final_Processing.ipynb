At this point, a major issue arises. FIFA stores player names in two ways: "short name", which is generally their first initial and last name, and "long name", which is their full name. Conversely, the Premier League data is stored as the player's first and last name. As such, there is no simple way to join the two datasets. For example, Cristiano Ronaldo's short name matches with the Premier League data, but his long name does not, while Marcus Rashford's short name is different, but his long name remains the same. 

As such, the easiest solution I could find was to convert every player's name into a "short name". For most players, this can be achieved through replacing the first word of their names with their first initial.

```python
name_temp = df_2122["Player"].str.split(" ")
final_names = []
for names in name_temp:
  names[0] = names[0][0:1] + "."
  final_names.append(" ".join(names))

df_2122["short_name"] = final_names
df_2122
```

At this point, I opened these files in Microsoft Excel, and text edited the "short_name" column to fix any differences between the FIFA "short_name" and the column I made for the Premier League data. Some example include names that are only one word, names in which multiple names are shortened, and names in which the dataset and FIFA order names differently (i.e. Heung-Min Son vs. Son Heung-Min).

After this, I reuploaded these edited files to my google drive, which will now be joined together.

For my purposes, I will be using an inner join, as I want to have full data for every player included

```python
df_prem_2122 = pd.read_csv(data_dir + "2122Prem.csv")
df_prem_2122 = df_prem_2122.drop_duplicates(subset=['Player'])
df_fifa_23 = pd.read_csv(data_dir + "23Prem.csv")
df_joined23 = df_prem_2122.merge(df_fifa_23, on="short_name")

df_prem_2021 = pd.read_csv(data_dir + "2021Prem.csv")
df_prem_2021 = df_prem_2021.drop_duplicates(subset=['Player'])
df_fifa_22 = pd.read_csv(data_dir + "22Prem.csv")
df_joined22 = df_prem_2021.merge(df_fifa_23, on="short_name")

df_prem_1920 = pd.read_csv(data_dir + "1920Prem.csv")
df_prem_1920 = df_prem_1920.drop_duplicates(subset=['Player'])
df_fifa_21 = pd.read_csv(data_dir + "21Prem.csv")
df_joined21 = df_prem_1920.merge(df_fifa_21, on="short_name")

df_prem_1819 = pd.read_csv(data_dir + "1819Prem.csv")
df_prem_1819 = df_prem_1819.drop_duplicates(subset=['Player'])
df_fifa_20 = pd.read_csv(data_dir + "20Prem.csv")
df_joined20 = df_prem_1819.merge(df_fifa_20, on="short_name")

df_prem_1718 = pd.read_csv(data_dir + "1718Prem.csv")
df_prem_1718 = df_prem_1718.drop_duplicates(subset=['Player'])
df_fifa_19 = pd.read_csv(data_dir + "19Prem.csv")
df_joined19 = df_prem_2021.merge(df_fifa_19, on="short_name")
```

```python
df_2122Defense = pd.read_csv(data_dir + "2021-22_Defense.txt")
df_2122Defense = df_2122Defense.drop_duplicates(subset=['Player'])
df_2122Defense["Pos"] = df_2122Defense["Pos"].str[:2]
df_2122Defense["Nation"] = df_2122Defense["Nation"].str[-3:]
df_2122Defense = df_2122Defense.iloc[:, :-2]
df_joined23 = df_joined23.merge(df_2122Defense, on=["Player", "Nation", "Age",
                                                    "Pos", "Squad", "90s"])

df_2021Defense = pd.read_csv(data_dir + "2020-21_Defense.txt")
df_2021Defense = df_2021Defense.drop_duplicates(subset=['Player'])
df_2021Defense["Pos"] = df_2021Defense["Pos"].str[:2]
df_2021Defense["Nation"] = df_2021Defense["Nation"].str[-3:]
df_2021Defense = df_2021Defense.iloc[:, :-2]
df_joined22 = df_joined22.merge(df_2021Defense, on=["Player", "Nation", "Age",
                                                    "Pos", "Squad", "90s"])

df_1920Defense = pd.read_csv(data_dir + "2019-20_Defense.txt")
df_1920Defense = df_1920Defense.drop_duplicates(subset=['Player'])
df_1920Defense["Pos"] = df_1920Defense["Pos"].str[:2]
df_1920Defense["Nation"] = df_1920Defense["Nation"].str[-3:]
df_1920Defense = df_1920Defense.iloc[:, :-2]
df_joined21 = df_joined21.merge(df_1920Defense, on=["Player", "Nation", "Age",
                                                    "Pos", "Squad", "90s"])

df_1819Defense = pd.read_csv(data_dir + "2018-19_Defense.txt")
df_1819Defense = df_1819Defense.drop_duplicates(subset=['Player'])
df_1819Defense["Pos"] = df_1819Defense["Pos"].str[:2]
df_1819Defense["Nation"] = df_1819Defense["Nation"].str[-3:]
df_1819Defense = df_1819Defense.iloc[:, :-2]
df_joined20 = df_joined20.merge(df_1819Defense, on=["Player", "Nation", "Age",
                                                    "Pos", "Squad", "90s"])

df_1718Defense = pd.read_csv(data_dir + "2017-18_Defense.txt")
df_1718Defense = df_1718Defense.drop_duplicates(subset=['Player'])
df_1718Defense["Pos"] = df_1718Defense["Pos"].str[:2]
df_1718Defense["Nation"] = df_1718Defense["Nation"].str[-3:]
df_1718Defense = df_1718Defense.iloc[:, :-2]
df_joined19 = df_joined19.merge(df_1718Defense, on=["Player", "Pos"], 
                          suffixes=("", ".1"))
```

```python
df_joined = pd.concat([df_joined22, df_joined23], ignore_index=True, axis=0)
df_joined = pd.concat([df_joined, df_joined21], ignore_index=True, axis=0)
df_joined = pd.concat([df_joined, df_joined20], ignore_index=True, axis=0)
df_joined = pd.concat([df_joined, df_joined19], ignore_index=True, axis=0)
df_joined = df_joined[df_joined["Pos"] != "GK"]
print(df_joined.columns.tolist())
df_joined=df_joined.iloc[:, 2:]
df_joined
```
